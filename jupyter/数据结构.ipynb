### 连续存储 -- 数组

```c
#include <stdio.h>
#include <time.h>
#include <stdlib.h>
#include <sys/malloc.h>

#define MAXSIZE 8

typedef int ElementType;

typedef struct LNode
{
    ElementType data[MAXSIZE];
    int Last;
} List_t, *pList_t;

pList_t initList();
void Insert(ElementType, int, pList_t);
void Delete(int, pList_t);

int main()
{
    pList_t L = initList();
    Insert(1, 1, L);
    Delete(1, L);
}

pList_t initList()
{
    pList_t ptrList = (pList_t)malloc(sizeof(List_t));
    ptrList->Last = -1;
    return ptrList;
}

int find(ElementType x, pList_t ptrList)
{
    int i = 0;
    while (i <= ptrList->Last && ptrList->data[i] != x)
        i++;
    if (i > ptrList->Last)
        return -1;
    else
        return i;
}

void Insert(ElementType x, int i, pList_t ptrList)
{
    int j;
    if (ptrList->Last == MAXSIZE - 1)
    {
        fprintf(stderr, "[Error::%s] The list is full\n", __func__);
        exit(-1);
    }
    if (i < 1 || i> ptrList->Last + 2)
    {
        fprintf(stderr, "[Error::%s] Invalid position!\n", __func__);
        exit(-1);
    }

    for (j = ptrList->Last; j >= i-1; j --)
        ptrList->data[j+1] = ptrList->data[j];
    ptrList->data[i-1] = x;
    ptrList->Last ++;
    return;
}

void Delete(int i, pList_t ptrList)
{
    int j;
    if (i < 1 || i> ptrList->Last + 1)
    {
        fprintf(stderr, "[Error::%s] Invalid position!\n", __func__);
        exit(-1);
    }
    for (j = i; j <ptrList->Last; j++)
        ptrList->data[j-1] = ptrList->data[j];
    ptrList->Last--;

    return;
}
```

#### 链表

示例

a. 创建一个非循环单链表

```c
#include <stdio.h>
#include <sys/malloc.h> //on Mac
#include <stdlib.h>
#include <stdbool.h>

typedef struct Node 
{
    int data; //数据域
    struct Node *pNext; //指针域
} NODE, *PNODE; // NODE 等价于struct Node; PNODE 等价于 struct Node *


PNODE create_list();


int main()
{
    PNODE pHead = NULL; //等价于 struct Node *Phead = NULL;
    pHead = create_list(); //创建一个非循环单链表
    return 0;
}


PNODE create_list()
{
    int len; //用来存储有效节点的个数
    int i; 
    int val; //用来临时存放用户输入的节点的值 
    
    PNODE pHead = (PNODE)malloc(sizeof(NODE)); //头结点
    if (NULL == pHead) 
    {
        fprintf(stderr, "[E::%s] Out of memory. Please check!", __func__);
        exit(-1);
    }
    
    PNODE pTail = pHead; //尾结点 
    pTail->pNext = NULL;
    
    
    printf("Please input the number of nodes you want to create: len= ");
    scanf("%d", &len);
    
    for (i = 0; i < len; i++)
    {
         printf("Please the data value for node: %d", i + 1);
         scanf("%d", &val);
         
         PNODE pNew = (PNODE)malloc(sizeof(NODE)); //新节点
         if (NULL == pNew)
         {
             fprintf(stderr, "[E::%s] Out of memory. Please check!", __func__);
             exit(-2);
         }
         pNew->data = val;
         pTail->pNext = pNew; 
         pNew->pNext = NULL;
         pTail = pNew;
    }
    
    return pHead;
}
```

b. 查看链表是否为空

```c
bool is_empty(PNODE pHead)
{
    if (NULL == pHead->pNext) //查看节点1
        return true;
    else
        return false;
}
```

c. 遍历链表

```c
void traverse_list(PNODE pHead)
{
    PNODE p = pHead->pNext; //节点1
    if (NULL == p) //节点1为空；也可以使用is_empty去检测
    {
        printf("The list is empty!\n");
        return;
    }
    
    while (NULL != p) //如果节点1不为空，则遍历余下节点
    {
        printf("%d ", p->data);
        p = p->pNext;
    }
    printf("\n");
    
    return;
}
```

d. 链表排序

这里可以按照数组排序的思路去排序链表（示例：冒泡排序）

```c
void sort_list(PNODE pHead)
{
    int i, j, t, len;
    
    PNODE p, q;
    
    len = length_list(pHead);
    
    for (i = 0, p = pHead->pNext; i < len -1; ++i, p = p->pNext)
    {
        for (j = i + 1, q = p->pNext; j < len; ++j, q = q->pNext)
        {
            if (p->data > q->data) //类似于数组中的： a[i] > a[j]
            {
                t = p->data;
                p->data = q->data;
                q->data = t;
            }           
        }
    }
    return;
}
```

e. 元素的插入

```c
bool insert_list(PNODE pHead, int pos, int val)
{
    int i = 0; 
    PNODE p = pHead;
    
    while (NULL != p && i < pos - 1)
    {
        p = p->pNext;
        ++i;
    }
    
    if (i > pos -1 || NULL == p)
    {
        return false;
    }
    
    PNODE pNew = (PNODE)malloc(sizeof(NODE));
    
    if (NULL == pNew)
    {
        fprintf(stderr, "[E::%s] Out of memory. Please check!", __func__);
        exit(-3);
    }
    
    pNew->data = val;
    PNODE q = p->pNext;
    p->pNext = pNew;
    pNew->pNext = q;
    
    return true;
}
```

f. 元素的删除

```c
bool delete_list(PNODE pHead, int pos, int * pVal)
{
    int i = 0; 
    PNODE p = pHead;
    
    while (NULL != p && i < pos - 1)
    {
        p = p->pNext;
        ++i;
    }
    
    if (i > pos -1 || NULL == p)
    {
        return false;
    }
    
    PNODE q = p->pNext; 
    *pVal = q->data;
    
    //删除p节点后面的节点
    p->pNext = p->pNext->pNext;
    free(q); 
    q = NULL;
    
    return true;
}
```

g. 链表的销毁

```c
void destory_list(PNODE pHead)
{
    PNODE p;
    if (NULL == pHead)
    {
        fprintf(stderr, "[E::%s] No data passed!", __func__);
        exit(-4);
    }

    while (pHead != NULL)
    {
        p = pHead->pNext;
        free(pHead);
        pHead = p;
    }
    return;
}
```

h. 完整代码

```c
#include <stdio.h>
#include <sys/malloc.h> //on Mac
#include <stdlib.h>
#include <stdbool.h>

typedef struct Node 
{
    int data; //数据域
    struct Node *pNext; //指针域
} NODE, *PNODE; // NODE 等价于struct Node; PNODE 等价于 struct Node *


PNODE create_list();
void traverse_list(PNODE);
bool is_empty(PNODE);
int length_list(PNODE);
bool insert_list(PNODE, int, int);
bool delete_list(PNODE, int, int *);
void sort_list(PNODE);
void destory_list(PNODE);

int main()
{   
    int val;
    PNODE pHead = NULL; //等价于 struct Node *Phead = NULL;
    pHead = create_list(); //创建一个非循环单链表
    traverse_list(pHead);
    
    insert_list(pHead, 3, 33);
    
    if(delete_list(pHead, 4, &val))
    {
        printf("Delete '%d' successfully!\n", val);
    }
    else
    {
        printf("Fail to delete '%d'!\n", val);
    }
    
    traverse_list(pHead);
    destory_list(pHead);
    
    return 0;
}

PNODE create_list()
{
    int len; //用来存储有效节点的个数
    int i; 
    int val; //用来临时存放用户输入的节点的值 
    
    PNODE pHead = (PNODE)malloc(sizeof(NODE)); //头结点
    if (NULL == pHead) 
    {
        fprintf(stderr, "[E::%s] Out of memory. Please check!", __func__);
        exit(-1);
    }
    
    PNODE pTail = pHead; //尾结点 
    pTail->pNext = NULL;
    
    
    printf("Please input the number of nodes you want to create: len = ");
    scanf("%d", &len);
    
    for (i = 0; i < len; i++)
    {
         printf("Please input data value for node: %d ", i + 1);
         scanf("%d", &val);
         
         PNODE pNew = (PNODE)malloc(sizeof(NODE)); //新节点
         if (NULL == pNew)
         {
             fprintf(stderr, "[E::%s] Out of memory. Please check!", __func__);
             exit(-2);
         }
         pNew->data = val;
         pTail->pNext = pNew; 
         pNew->pNext = NULL;
         pTail = pNew;
    }
    
    return pHead;
}

void traverse_list(PNODE pHead)
{
    PNODE p = pHead->pNext;
    while (NULL != p)
    {
        printf("%d ", p->data);
        p = p->pNext;
    }
    printf("\n");
    
    return;
}


bool is_empty(PNODE pHead)
{
    if (NULL == pHead->pNext)
        return true;
    else
        return false;
}

int length_list(PNODE pHead)
{
    PNODE p = pHead->pNext;
    int len = 0;
    
    while (NULL != p)
    {
        ++len;
        p = p->pNext;
    }
    return len;
}


void sort_list(PNODE pHead)
{
    int i, j, t, len;
    
    PNODE p, q;
    
    len = length_list(pHead);
    
    for (i = 0, p = pHead->pNext; i < len -1; ++i, p = p->pNext)
    {
        for (j = i + 1, q = p->pNext; j < len; ++j, q = q->pNext)
        {
            if (p->data > q->data) //类似于数组中的： a[i] > a[j]
            {
                t = p->data;
                p->data = q->data;
                q->data = t;
            }           
        }
    }
    return;
}

bool insert_list(PNODE pHead, int pos, int val)
{
    int i = 0; 
    PNODE p = pHead;
    
    while (NULL != p && i < pos - 1)
    {
        p = p->pNext;
        ++i;
    }
    
    if (i > pos -1 || NULL == p)
    {
        return false;
    }
    
    PNODE pNew = (PNODE)malloc(sizeof(NODE));
    
    if (NULL == pNew)
    {
        fprintf(stderr, "[E::%s] Out of memory. Please check!", __func__);
        exit(-3);
    }
    
    pNew->data = val;
    PNODE q = p->pNext;
    p->pNext = pNew;
    pNew->pNext = q;
    
    return true;
}


bool delete_list(PNODE pHead, int pos, int * pVal)
{
    int i = 0; 
    PNODE p = pHead;
    
    while (NULL != p && i < pos - 1)
    {
        p = p->pNext;
        ++i;
    }
    
    if (i > pos -1 || NULL == p)
    {
        return false;
    }
    
    PNODE q = p->pNext; 
    *pVal = q->data;
    
    //删除p节点后面的节点
    p->pNext = p->pNext->pNext;
    free(q); 
    q = NULL;
    
    return true;
}

void destory_list(PNODE pHead)
{
    PNODE p;
    if (NULL == pHead)
    {
        fprintf(stderr, "[E::%s] No data passed!", __func__);
        exit(-4);
    }

    while (pHead != NULL)
    {
        p = pHead->pNext;
        free(pHead);
        pHead = p;
    }
    return;
}
```
